```python
#svm
from sklearn import svm 
```

```python
#librairies pour la visualisation de donn√©es
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(font_scale=1.2)
```

```python
# plot 2 neuronnes
sns.lmplot('AF3' ,'AF4' ,data=dataset,hue='eyeDetection',fit_reg=False,scatter_kws={"s":20});
```

```python
#specify inputs for the model
features= dataset[['AF3','AF4']].values
type_label = np.where(dataset['eyeDetection']==1,0,1)
```

```python
#fit the SVM model
model = svm.SVC(kernel='linear')
#apprentissage
model.fit(features,type_label)
```

```python
#C : float, optionnel (par d√©faut = 1.0)
#Param√®tre de p√©nalit√© C du terme d'erreur.

#kernel : cha√Æne de caract√®res, optionnel (default = 'rbf')
#Sp√©cifie le type de noyau √† utiliser dans l'algorithme dans notre cas est lin√©aire

#degr√© : int, facultatif (par d√©faut = 3)
#Degr√© de la fonction du noyau polynomiale ('poly')

#gamma : float, optionnel 
#Coefficient de noyau pour 'rbf', 'poly' et 'sigmo√Øde'.

#coef0 : float, facultatif (par d√©faut = 0.0)
#Terme ind√©pendant dans la fonction du noyau.

#shrinking : boolean, facultatif (par d√©faut = True)
#S'il faut utiliser l'heuristique de r√©duction.

#probabilit√© : bool√©en, optionnel (par d√©faut = False)
#Activer ou non les estimations de probabilit√©. 

#tol : float, optionnel (par d√©faut = 1e-3)
#Tol√©rance pour le crit√®re d'arr√™t.

#taille_cache : float, optionnel
#Sp√©cifiez la taille du cache du noyau (en Mo).

#class_weight : {dict, 'balanc√©'}, facultatif
#D√©finissez le param√®tre C de la classe i sur class_weight [i] * C pour SVC. Si ce n'est pas donn√©, toutes les classes sont suppos√©es avoir un poids un. Le mode ¬´√©quilibr√©¬ª utilise les valeurs de y pour ajuster automatiquement les poids inversement proportionnels aux fr√©quences de classe dans les donn√©es d‚Äôentr√©e.n_samples / (n_classes * np.bincount(y))

#verbose : bool, default: False
#Activer la sortie comment√©e.

#max_iter : int, facultatif (par d√©faut = -1)
#Limite stricte sur les it√©rations dans le solveur, ou -1 pour aucune limite.

#decision_function_shape : 'ovo', 'ovr', default = 'ovr'
#Indique s'il faut renvoyer une fonction de d√©cision one-vs-rest ('ovr') de forme (n_samples, n_classes) comme tous les autres classificateurs, ou la fonction de d√©cision one-vs-one ('ovo') d'origine de libsvm qui a la forme (n_samples , n_classes * (n_classes - 1) / 2). Cependant, one-vs-one ('ovo') est toujours utilis√© comme strat√©gie multi-classes.
#random_seed : int, optionnel
#Graine pour le g√©n√©rateur de nombres al√©atoires utilis√© pour les estimations de probabilit√©. 0 par d√©faut."""

```

```python
#get the separating hyperplane
w = model.coef_[0] 
a = -w[0] / w[1]
xx = np.linspace(0,6000)
yy = a*xx - (model.intercept_[0]) / w[1]
```

```python
#plot the parallels to the  separating hyperplane that pass through the support vectors
b= model.support_vectors_[0]
yy_down = a*xx + (b[1] - a*b[0])
b= model.support_vectors_[-1]
yy_up = a*xx + (b[1] - a*b[0])
```

```python
# Plot the hyperplane
sns.lmplot('AF3','AF4' ,data=dataset,hue='eyeDetection',palette='Set1',fit_reg=False,scatter_kws={"s":70})
plt.plot(xx,yy, linewidth=2, color='black');
```

```python
#look at the margins and support vectors

sns.lmplot('AF3','AF4' ,data=dataset,hue='eyeDetection',palette='Set1',fit_reg=False,scatter_kws={"s":70})
plt.plot(xx,yy, linewidth=2, color='black')
plt.plot(xx , yy_down,'k--')
plt.plot(xx , yy_up,'k--')
plt.scatter(model.support_vectors_[:,1],model.support_vectors_[:,1],model.support_vectors_[:,1],s=80,facecolors='none')
```

```python
#create this function to guess when the eye is closed or opened
def eyedetection(AF3,AF4):
    if(model.predict([[AF3,AF4]]))==0:
        print('opened')
    else:
        print('closed')
```

```python
#predict if 2000 AF3  and AF4 the eye state 
eyedetection(4324.62,4384.1)
```

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```
