```python
#svm
from sklearn import svm 
```

```python
#librairies pour la visualisation de données
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(font_scale=1.2)
```

```python
# plot 2 neuronnes
sns.lmplot('AF3' ,'AF4' ,data=dataset,hue='eyeDetection',fit_reg=False,scatter_kws={"s":20});
```

```python
#specify inputs for the model
features= dataset[['AF3','AF4']].values
type_label = np.where(dataset['eyeDetection']==1,0,1)
```

```python
#fit the SVM model
model = svm.SVC(kernel='linear')
#apprentissage
model.fit(features,type_label)
```

```python
#C : float, optionnel (par défaut = 1.0)
#Paramètre de pénalité C du terme d'erreur.

#kernel : chaîne de caractères, optionnel (default = 'rbf')
#Spécifie le type de noyau à utiliser dans l'algorithme dans notre cas est linéaire

#degré : int, facultatif (par défaut = 3)
#Degré de la fonction du noyau polynomiale ('poly')

#gamma : float, optionnel 
#Coefficient de noyau pour 'rbf', 'poly' et 'sigmoïde'.

#coef0 : float, facultatif (par défaut = 0.0)
#Terme indépendant dans la fonction du noyau.

#shrinking : boolean, facultatif (par défaut = True)
#S'il faut utiliser l'heuristique de réduction.

#probabilité : booléen, optionnel (par défaut = False)
#Activer ou non les estimations de probabilité. 

#tol : float, optionnel (par défaut = 1e-3)
#Tolérance pour le critère d'arrêt.

#taille_cache : float, optionnel
#Spécifiez la taille du cache du noyau (en Mo).

#class_weight : {dict, 'balancé'}, facultatif
#Définissez le paramètre C de la classe i sur class_weight [i] * C pour SVC. Si ce n'est pas donné, toutes les classes sont supposées avoir un poids un. Le mode «équilibré» utilise les valeurs de y pour ajuster automatiquement les poids inversement proportionnels aux fréquences de classe dans les données d’entrée.n_samples / (n_classes * np.bincount(y))

#verbose : bool, default: False
#Activer la sortie commentée.

#max_iter : int, facultatif (par défaut = -1)
#Limite stricte sur les itérations dans le solveur, ou -1 pour aucune limite.

#decision_function_shape : 'ovo', 'ovr', default = 'ovr'
#Indique s'il faut renvoyer une fonction de décision one-vs-rest ('ovr') de forme (n_samples, n_classes) comme tous les autres classificateurs, ou la fonction de décision one-vs-one ('ovo') d'origine de libsvm qui a la forme (n_samples , n_classes * (n_classes - 1) / 2). Cependant, one-vs-one ('ovo') est toujours utilisé comme stratégie multi-classes.
#random_seed : int, optionnel
#Graine pour le générateur de nombres aléatoires utilisé pour les estimations de probabilité. 0 par défaut."""

```

```python
#get the separating hyperplane
w = model.coef_[0] 
a = -w[0] / w[1]
xx = np.linspace(0,6000)
yy = a*xx - (model.intercept_[0]) / w[1]
```

```python
#plot the parallels to the  separating hyperplane that pass through the support vectors
b= model.support_vectors_[0]
yy_down = a*xx + (b[1] - a*b[0])
b= model.support_vectors_[-1]
yy_up = a*xx + (b[1] - a*b[0])
```

```python
# Plot the hyperplane
sns.lmplot('AF3','AF4' ,data=dataset,hue='eyeDetection',palette='Set1',fit_reg=False,scatter_kws={"s":70})
plt.plot(xx,yy, linewidth=2, color='black');
```

```python
#look at the margins and support vectors

sns.lmplot('AF3','AF4' ,data=dataset,hue='eyeDetection',palette='Set1',fit_reg=False,scatter_kws={"s":70})
plt.plot(xx,yy, linewidth=2, color='black')
plt.plot(xx , yy_down,'k--')
plt.plot(xx , yy_up,'k--')
plt.scatter(model.support_vectors_[:,1],model.support_vectors_[:,1],model.support_vectors_[:,1],s=80,facecolors='none')
```

```python
#create this function to guess when the eye is closed or opened
def eyedetection(AF3,AF4):
    if(model.predict([[AF3,AF4]]))==0:
        print('opened')
    else:
        print('closed')
```

```python
#predict if 2000 AF3  and AF4 the eye state 
eyedetection(4324.62,4384.1)
```
